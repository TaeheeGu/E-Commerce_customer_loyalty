```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./Dataset/E-commerce(visualization).csv")
print(df.columns)
```

### category에 따른 분포

```python
category_series = df["category"].value_counts(sort=True, normalize=True)
plt.barh(category_series.index,category_series,color="green",linewidth=1.5)
plt.title('카테고리별 구매량')
plt.show()
```

```python
category_series = df["category"].value_counts(sort=True, normalize=True)
bar2 = plt.barh(category_series.index,category_series,color="green",linewidth=1.5,)

weeknd = df[(df["dayofweek"] == "Sat") | (df["dayofweek"] == "Sun")]
category_series = weeknd["category"].value_counts(sort=True)/6080761
bar1 = plt.barh(category_series.index,category_series,color="orange",linewidth=1.5)

plt.title('카테고리별 구매량 (주말)')
plt.legend((bar1, bar2), ('주말', '평일'), fontsize=10)
plt.show()
```

### category_construction

```python
construction = df.loc[df["category"]== "construction"]
event_days = (construction["dayofweek"].value_counts(sort=False, normalize=True))

ymin = event_days.min()
ymax = event_days.max()

plt.plot(event_days,color="orange",linewidth=2)
plt.title('건설 용품 : 요일에 따른 사용자의 구매 그래프')
plt.xlim(0,6)
plt.ylim(0.12, ymax+0.02)
plt.grid(True)
plt.fill_between(event_days.index,event_days,color="orange",alpha=0.1)
plt.show()
print(event_days*100)
```

### event_hour에 따른 분포

```python
event_hours = (df["event_hour"].value_counts(sort=False))

ymin = event_hours.min()
ymax = event_hours.max()

plt.plot(event_hours,color="green",linewidth=1.5)
plt.title('시간에 따른 사용자의 구매 그래프')
plt.xlim(0,23)
plt.ylim(ymin, ymax+50000)
plt.grid(True)

plt.fill_between(event_hours.index[6:13],event_hours[6:13],color="green",alpha=0.3)
plt.fill_between(event_hours.index[12:19],event_hours[12:19],color="green",alpha=0.2)
plt.show()

# 아침부터 오후시간대까지가 가장 많다. 총 6080761
print("오전 6부터 오전 12까지 구매한 고객의 비율 : ",df["event_hour"].value_counts(sort=False, normalize=True)[4:11].sum()*100)
print("오전 12부터 오후 6까지 구매한 고객의 비율 : ",df["event_hour"].value_counts(sort=False, normalize=True)[10:17].sum()*100)
```

### weekend에 따른 분포

```python
# 주말 판매시간
weeknd = df[(df["dayofweek"] == "Sat") | (df["dayofweek"] == "Sun")]
event_hour_weeknd = weeknd["event_hour"].value_counts(sort=False)
ymin = event_hour_weeknd.min()
ymax = event_hour_weeknd.max()

plt.plot(event_hour_weeknd,color="green",linewidth=1.5)
plt.title('이용 시간에 따른 사용자의 구매 빈도수 (주말)')
plt.xlim(0,23)
plt.ylim(ymin, 500615)
plt.grid(True)
plt.fill_between(event_hour_weeknd.index[6:13],event_hour_weeknd[6:13],color="green",alpha=0.3)
plt.fill_between(event_hour_weeknd.index[12:19],event_hour_weeknd[12:19],color="green",alpha=0.2)
plt.show()

print("오전 6부터 오전 12까지 구매한 고객의 비율 : ",weeknd["event_hour"].value_counts(sort=False, normalize=True)[4:11].sum()*100)
print("오전 12부터 오후 6까지 구매한 고객의 비율 : ",weeknd["event_hour"].value_counts(sort=False, normalize=True)[10:17].sum()*100)
```

### weekday에 따른 분포

```python
# 평일 판매시간
weekday = df[~((df["dayofweek"] == "Sat") | (df["dayofweek"] == "Sun"))]

event_hour_weekday = weekday["event_hour"].value_counts(sort=False)
ymin = event_hour_weekday.min()
ymax = event_hour_weekday.max()

plt.plot(event_hour_weekday,color="green",linewidth=1.5)
plt.ylim(ymin, 500615)
plt.grid(True)
plt.xlim(0,23)
plt.grid(True)

plt.fill_between(event_hour_weekday.index[6:13],event_hour_weekday[6:13],color="green",alpha=0.3)
plt.fill_between(event_hour_weekday.index[12:19],event_hour_weekday[12:19],color="green",alpha=0.2)
plt.title('이용 시간에 따른 사용자의 구매 빈도수 (평일)')
plt.show()

print("오전 6부터 오전 12까지 구매한 고객의 비율 : ",weekday["event_hour"].value_counts(sort=False, normalize=True)[4:11].sum()*100)
print("오전 12부터 오후 6까지 구매한 고객의 비율 : ",weekday["event_hour"].value_counts(sort=False, normalize=True)[10:17].sum()*100)
```

### weekend vs weekday

```python
weekday = df[~((df["dayofweek"] == "Sat") | (df["dayofweek"] == "Sat"))]
event_hour_weekday = (weekday["event_hour"].value_counts(sort=False))
ymin = event_hour_weekday.min()
ymax = event_hour_weekday.max()
plt.plot(event_hour_weekday,color="green",linewidth=1.5)
plt.fill_between(event_hour_weekday.index[6:13],event_hour_weekday[6:13],color="green",alpha=0.3)
plt.fill_between(event_hour_weekday.index[12:19],event_hour_weekday[12:19],color="green",alpha=0.2)
plt.grid(True)

weeknd = df[(df["dayofweek"] == "Sat") | (df["dayofweek"] == "Sun")]
event_hour_weeknd = (weeknd["event_hour"].value_counts(sort=False))
ymin = event_hour_weeknd.min()
ymax = event_hour_weeknd.max()
plt.plot(event_hour_weeknd,color="green",linewidth=1.5)
plt.title('이용 시간에 따른 사용자의 구매 빈도수 (평일 vs 주말)')
plt.xlim(0,23)
plt.ylim(ymin-10000, 500615)
plt.grid(True)
plt.show()
```
