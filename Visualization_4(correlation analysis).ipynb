```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#데이터 불러오기
df = pd.read_csv("./Dataset/E-commerce(royalty).csv")
```

```python
df.columns
```

### 피어슨 상관계수

```python
#pearson 상관계수
pearson_corr = df.corr(method='pearson')
pearson_corr
```

```python
#상관계수 히트맵
plt.figure(figsize=(20,20))
pearson_heatmap = sns.heatmap(pearson_corr, cbar = True, annot = True, annot_kws={'size' : 10},
                                                     fmt = '.3f', square = True, cmap = 'Blues')
```

```python
s = pearson_corr['customer_royalty']

# Series이므로 DataFrame으로 변경한다. 
pearson_df = pd.DataFrame(s[s < 1].sort_values(ascending=False))
pearson_df.style.background_gradient(cmap='viridis')
```

### 스피어만 상관계수

```python
#spearman 상관계수
spearman_corr = df.corr(method='spearman')
spearman_corr
```

```python
#상관계수 히트맵
plt.figure(figsize=(20,20))
spearman_heatmap = sns.heatmap(spearman_corr, cbar = True, annot = True, annot_kws={'size' : 10},
                         fmt = '.3f', square = True, cmap = 'Blues')
```

```python
s = spearman_corr['customer_royalty']

# Series이므로 DataFrame으로 변경한다. 
spearman_df = pd.DataFrame(s[s < 1].sort_values(ascending=False))
spearman_df.style.background_gradient(cmap='viridis')
```

### 켄달타우 상관계수

```python
#kendalltau 상관계수
kendalltau_corr = df.corr(method='kendall')
kendalltau_corr
```

```python
#상관계수 히트맵
plt.figure(figsize=(20,20))
spearman_heatmap = sns.heatmap(kendalltau_corr, cbar = True, annot = True, annot_kws={'size' : 10},
                         fmt = '.3f', square = True, cmap = 'Blues')
```

```python
s = kendalltau_corr['customer_royalty']

# Series이므로 DataFrame으로 변경한다. 
kendalltau_df = pd.DataFrame(s[s < 1].sort_values(ascending=False))
kendalltau_df.style.background_gradient(cmap='viridis')
```
