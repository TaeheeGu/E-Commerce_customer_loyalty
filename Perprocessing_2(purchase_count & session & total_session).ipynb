```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#데이터 불러오기
df = pd.read_csv("./Dataset/E-commerce(one-hot).csv")
print(df.columns)
df.head()
```

```python
print(len(df))
```

# 사용자의 접속(user_session)별 구매

### purchase_count

```python
df['purchase_count'] = 1
```

```python
group = df.groupby(['user_session', 'user_id'], as_index = False)
#user_id, user_session로 각 유저(user_id)에 대해 동일한 접속기록(user_session)별 구매 기록 합계
user_df = group.sum()
```

```python
#각 세션당 구매이벤트 병합 한 세션당 하나의 row
user_df[['user_id', 'user_session', 'purchase_count']]
```

### session

```python
#유저별 접속 횟수를 확인
session_df = df[['user_id', 'user_session', 'event_time']]
#유저별 접속 시간에 따라 정렬
session_df = session_df.sort_values(["user_id", "event_time"], axis=0)
#접속 횟수를 확인하기 위해 중복되는 user_session 삭제
session_df = session_df.drop_duplicates(['user_session'], keep = 'first', ignore_index = True)

#user_id, user_session, event_time, session, total_session 세션 구분을 위한 데이터프레임
session_df.head()
```

```python
#각 세션별 시간순서를 통한 카운트
session_count_df = session_df[['user_id', 'user_session']]
 #동일 유저에 대한 각각의 접속, event_time을 통해 접속 기록 확인
session_count_df['session'] = 0
session_count_df.head()
```

```python
#빠른 연산을 위해 리스트로 변환
session_list = session_count_df.values.tolist()
#하나의 유저의 접속 횟수(순서)
count = 0
for i in range(1, len(session_list)):
    if session_list[i-1][0] == session_list[i][0]:
        count += 1
        session_list[i][2] = count
    else:
        count = 0

print(session_list[0:10])

#리스트 연산후 데이터프레임으로 변환
cols = ['user_id', 'user_session', 'session']
session_count_df = pd.DataFrame(session_list, columns = cols)
# 유저의 각각의 접속에 따른 접속 횟수
session_count_df.head()
```

### total_session

```python
#total_session을 구하기위한 total_session_df
total_session_df = session_df[['user_id']]
#유저의 총 접속 횟수 groupby의 sum을 활용하기 위해 초기 값 1 이후 재 조정
total_session_df['total_session'] = 1

#groupby를 통해 한 아이디 당 전체 세션 수를 구한다
grouped = total_session_df.groupby('user_id')
total_session_df = pd.DataFrame(grouped.sum()['total_session'])
total_session_df.reset_index(inplace = True)

#초기값 0부터 세션의 수를 세기 위해 -1
total_session_df['total_session'] = total_session_df['total_session'].apply(lambda x: x-1)

total_session_df.head()
```

### session & toteal_session

```python
final_session_df = pd.merge(session_count_df, total_session_df)
final_session_df.head()
```

### purchase_count & session & total_session merge

```python
#각 세션별 구매 합계 user_df와 유저별 세션 접속 카운트 session_count_df 를 공유키 user_id,user_session으로 병합
new_df = pd.merge(user_df, final_session_df)
new_df = new_df.sort_values(["user_id", "session"], axis=0) #user_id와 session 수로 정렬
new_df.drop(columns=["user_session", "user_id"],inplace=True) #user_session 컬럼 삭제
new_df.reset_index(drop = True, inplace = True) #인덱스 초기화
new_df.head()
```

```python
new_df.to_csv("./Dataset/E-commerce(split_session).csv", index=False)
```
